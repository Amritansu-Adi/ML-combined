```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
rng.random((5,)).shape
```

```python
class LinearRegression:
    def __init__(self, m):
        self.W = np.random.randn(m)
        self.b = np.random.randn()
    def forward(self, x):
        return x @ self.W + self.b
    def backward(self, loss, alpha, x):
        n = loss.shape[0]
        self.b -= alpha * 2/n * np.sum(loss) 
        self.W -= alpha * 2/n * x.T @ loss
```

```python
x = np.array([[1,2,3,4,5,6], [6,5,4,3,2,1]])
```

```python
y = np.array([0, 1])
model = LinearRegression(x.shape[1])
```

```python
epoch = 200
for _ in range(epoch):
    y_pred = model.forward(x)
    loss = (y_pred - y)
    model.backward(loss, 0.01, x)
```

```python
print(int(model.forward([1,2,4,50,6, 100])))
```
